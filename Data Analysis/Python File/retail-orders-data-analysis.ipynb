```python
#import libraries
import pandas as pd
```

```python
#read data from the file and checking# null values
df = pd.read_csv('/kaggle/input/retail-orders/orders.csv')
df.head(20)
```

```python
#finding out how many types of values present to figure out which values are null
df['Ship Mode'].unique()
```

```python
#converting not available and unknown types into null values (nan)
df = pd.read_csv('/kaggle/input/retail-orders/orders.csv',na_values=['Not Available','unknown'])
df['Ship Mode'].unique()
```

```python
#rename column names and make them lower case and replace space with underscores

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.head(20)
```

```python
#derive new columns discount, sale price and profit
df['discount_amount']=df['list_price']*df['discount_percent']*0.01
df['sale_price']=df['list_price']-(df['list_price']*df['discount_percent']*0.01)
df['profit']= df['sale_price']-df['cost_price']
df.head(10)
```

```python
#checking data types of our columns
df.dtypes
```

```python
#converting order date from object data type to datetime
df['order_date']=pd.to_datetime(df['order_date'],format='%Y-%m-%d')
df.dtypes
```

```python
#drop cost price list and discount percent columns
df=df.drop(columns=['list_price','discount_percent'])
df
```

```python
#saving the dataframe as updated csv file
df.to_csv('/kaggle/working/updated_retail_orders.csv', index=False)
```
